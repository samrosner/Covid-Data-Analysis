```python
import sys
print(sys.version)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.api import VAR
%matplotlib inline
```

```python
caseData=pd.read_csv("covidcases.csv")
caseData=caseData[caseData["Province_State"]=="Virginia"]
caseData=caseData.iloc[:,5:]
caseData=caseData.drop(["Province_State","Country_Region","Lat","Long_","Combined_Key"], axis=1)
caseData=caseData.set_index("Admin2")
caseData=caseData.T
caseData=caseData.iloc[:,1:10]

#try seven day rolling average
plt.plot(caseData["Albemarle"].diff().dropna()[50:150])
plt.plot(caseData["Albemarle"].diff().dropna()[50:150].rolling(7).mean().fillna(0))
plt.show()

trainData=caseData[50:400].diff().dropna().diff().dropna().rolling(7).mean().fillna(0)
testData=caseData[400:].diff().dropna().diff().dropna().rolling(7).mean().fillna(0)

print(testData)
```

```python
model = VAR(trainData)
model_fit=model.fit()
```

```python
lagged_values=trainData.values[-20:]
forecast=model_fit.forecast(y=lagged_values, steps=10)
albem=[]
for x in forecast:
    albem.append(x[1])
plt.plot(albem)
plt.plot(testData["Alexandria"][:10])
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
import math

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
trainKeras=caseData[50:400].values.astype('float32')

testKeras=caseData[400:].values.astype('float32')

#trainKeras=caseData[50:400].diff().dropna().diff().dropna().values.astype('float32')

#testKeras=caseData[400:].diff().dropna().diff().dropna().values.astype('float32')
```

```python
albemarleTrain=np.array([])

for x in trainKeras:
    albemarleTrain=np.append(albemarleTrain,x[0])
albemarleTrain=albemarleTrain.reshape((albemarleTrain.size,1))
scaler=MinMaxScaler(feature_range=(0,1))
albemarleTrain=scaler.fit_transform(albemarleTrain)

albemarleTest=np.array([])

for x in testKeras:
    albemarleTest=np.append(albemarleTest,x[0])
albemarleTest=albemarleTest.reshape((albemarleTest.size,1))
scaler=MinMaxScaler(feature_range=(0,1))
albemarleTest=scaler.fit_transform(albemarleTest)
```

```python
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back), 0]
        dataX.append(a)
        dataY.append(dataset[i + look_back, 0])
    return np.array(dataX), np.array(dataY)
```

```python
look_back=7
trainKerasX, trainKerasY = create_dataset(albemarleTrain, look_back)
testKerasX, testKerasY = create_dataset(albemarleTest, look_back)
```

```python
trainKerasX = np.reshape(trainKerasX, (trainKerasX.shape[0], 1, trainKerasX.shape[1]))
testKerasX = np.reshape(testKerasX, (testKerasX.shape[0], 1, testKerasX.shape[1]))
plt.plot(trainKerasX[:,0,0])
plt.plot(trainKerasX[:,0,6])
plt.show()
print(trainKerasX)
```

```python
trainKerasX[:-10,0,4:7]
```

```python
model = Sequential()
model.add(LSTM(4, input_shape=(1, look_back)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(trainKerasX, trainKerasY, epochs=100, batch_size=1, verbose=2)
```

```python

trainPredict = model.predict(trainKerasX)
testPredict = model.predict(testKerasX)

trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform([trainKerasY])
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform([testKerasY])
#print(trainPredict)

trainScore = math.sqrt(mean_squared_error(trainKerasY, trainPredict[:,0]))
print('Train Score: %.2f RMSE' % (trainScore))
testScore = math.sqrt(mean_squared_error(testKerasY, testPredict[:,0]))
print('Test Score: %.2f RMSE' % (testScore))
```
